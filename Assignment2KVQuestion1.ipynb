```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier

#importing the database over
df = pd.read_csv("https://github.com/mpourhoma/CS4661/raw/master/Heart_s2.csv")
```

```python
#This is just to see a portion of it for myself :))

#So I know the different features shown and how the data per feature looks like!
df[0:5]
```

```python
df.columns
```

```python
X = df[['Age', 'RestBP', 'Chol', 'MaxHR','Oldpeak']]
y = df['AHD']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=9)

#To see if code worked
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=9)
print(X_train.shape)
print(X_test.shape)
```

```python
#KNN k=7
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(X_train, y_train)
knnPred = knn.predict(X_test)
print("Accuracy of Knn = 7:", accuracy_score(y_test, knnPred))
```

```python
#Decision Tree w/ random_state=5
dTree = DecisionTreeClassifier(random_state=5)
dTree.fit(X_train, y_train)
dTreePred = dTree.predict(X_test)
print("Decision Tree with random state 5, Accuracy:", accuracy_score(y_test, dTreePred))
```

```python
#Logistic Regression w/ max_iter=400
logR = LogisticRegression(max_iter=400)
logR.fit(X_train, y_train)
logRPred = logR.predict(X_test)
print("Logistic with Max iteration 400 Accuracy:", accuracy_score(y_test, logRPred))
```

The best close to accuracy was KNN
