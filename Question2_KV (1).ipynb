# Part A
In the code below is the set up, from importing the Iris datafram from sklearn. 
To ensure it was done properly we print out the heads which will display the features with some rows of data.

Then here I set the label to y, and printing it to ensure I got the label vector right!

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score


iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

df["species"] = iris.target

y = iris.target

print(df.head())
print(y)
```

# Part B
For this section using the data we imported and "train_test_split" we want 0.4 of that data set as the test size and we use random state 10 so every time it is run we do not get newly randomly selected data every time!

I looked up the syntax on 
https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

And to ensure we have the right split I printed them to display the shape :)


```python
fTrain,fTest,lTrain,lTest = train_test_split(df,y, test_size=0.4,random_state=42)

print(fTrain.shape)
print(fTest.shape)
```

# Part C & D
By creating an array with the numbers we want k to be we could loop through them to print out the different results and compare them! :)) 

```python
num = [1,3,4,11,15,27,43]

for k in num:
    knn = KNeighborsClassifier(n_neighbors=k) 
    knn.fit(fTrain,lTrain)
    pred = knn.predict(fTest)
    print("k:", k)
    print(pred)
    accuracy = accuracy_score(lTest, pred)
    print(accuracy)
```

# Part E

My main focus wasn't to keep it clean I just wanted to ensure the code ran! And! I was able to individually get the single features and based on the single features train the data and test the data spliting it all the same way and then print the accuracy 

```python
f1 = df[['sepal length (cm)']] 
f1Train, f1Test, l1Train, l1Test = train_test_split(f1,y, test_size=0.4,random_state=42)
f1knn = KNeighborsClassifier(n_neighbors=11) 
f1knn.fit(f1Train,l1Train)
f1pred = f1knn.predict(f1Test)
print("Sepal Length")
print(f1pred)
f1accuracy = accuracy_score(l1Test, f1pred)
print(f1accuracy)

f2 = df[['sepal width (cm)']]
f2Train, f2Test, l2Train, l2Test = train_test_split(f2,y, test_size=0.4,random_state=42)
f2knn = KNeighborsClassifier(n_neighbors=11) 
f2knn.fit(f2Train,l2Train)
f2pred = f2knn.predict(f2Test)
print("Sepal width")
print(f2pred)
f2accuracy = accuracy_score(l2Test, f2pred)
print(f2accuracy)

f3 = df[['petal length (cm)']]
f3Train, f3Test, l3Train, l3Test = train_test_split(f3,y, test_size=0.4,random_state=42)
f3knn = KNeighborsClassifier(n_neighbors=11) 
f3knn.fit(f3Train,l3Train)
f3pred = f3knn.predict(f3Test)
print("petal Length")
print(f3pred)
f3accuracy = accuracy_score(l3Test, f3pred)
print(f3accuracy)

f4 = df[['petal width (cm)']]
f4Train, f4Test, l4Train, l4Test = train_test_split(f4,y, test_size=0.4,random_state=42)
f4knn = KNeighborsClassifier(n_neighbors=11) 
f4knn.fit(f4Train,l4Train)
f4pred = f4knn.predict(f4Test)
print("Petal Width")
print(f4pred)
f4accuracy = accuracy_score(l4Test, f4pred)
print(f4accuracy)
#print(f1)
```
